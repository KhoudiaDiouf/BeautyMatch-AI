```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Charger les données (remplacez le chemin par celui de votre fichier CSV)
df = pd.read_csv('cosmetics.csv')

# 1. Filtrage des produits par type de peau et budget

# Fonction pour filtrer les produits selon le type de peau et le budget
def filter_products(df, skin_type, budget):
    # Filtrer par type de peau
    if skin_type == 'dry':
        filtered_df = df[df['Dry'] == 1]
    elif skin_type == 'oily':
        filtered_df = df[df['Oily'] == 1]
    elif skin_type == 'combination':
        filtered_df = df[df['Combination'] == 1]
    elif skin_type == 'sensitive':
        filtered_df = df[df['Sensitive'] == 1]
    else:
        filtered_df = df  # Si aucun type de peau spécifique n'est choisi

    # Filtrer par budget
    if budget == 'economical':
        filtered_df = filtered_df[filtered_df['Price'] < 50]
    elif budget == 'standard':
        filtered_df = filtered_df[(filtered_df['Price'] >= 50) & (filtered_df['Price'] <= 150)]
    elif budget == 'premium':
        filtered_df = filtered_df[filtered_df['Price'] > 150]

    return filtered_df
```

```python
# Exemple d'utilisation du filtre : produit pour peau sèche et budget premium
skin_type = 'dry'
budget = 'premium'
filtered_products = filter_products(df, skin_type, budget)

# Affichage des produits filtrés
print(filtered_products[['Name', 'Price', 'Brand', 'Rank', 'Ingredients']])

# 2. Vectorisation des descriptions de produits avec TF-IDF

# Initialisation du vectoriseur TF-IDF
tfidf = TfidfVectorizer(stop_words='english')

# Vectorisation des descriptions de produits
tfidf_matrix = tfidf.fit_transform(df['Ingredients'])
```

```python
# 3. Calcul de la similarité cosinus pour trouver les produits similaires

# Fonction pour recommander des produits similaires à un produit donné
def recommend_similar_products(product_name, tfidf_matrix, top_n=5):
    # Trouver l'indice du produit donné
    product_idx = df[df['Name'] == product_name].index[0]
    
    # Calcul de la similarité cosinus entre le produit donné et tous les autres produits
    cosine_sim = cosine_similarity(tfidf_matrix[product_idx:product_idx+1], tfidf_matrix).flatten()
    
    # Récupérer les indices des produits les plus similaires
    similar_indices = cosine_sim.argsort()[-top_n-1:-1][::-1]
    
    # Récupérer les produits similaires
    similar_products = df.iloc[similar_indices]
    
    return similar_products[['Name', 'Price', 'Brand', 'Rank', 'Ingredients']]
```

```python
# Exemple de recommandation de produits similaires
product_name = 'Crème de la Mer'
similar_products = recommend_similar_products(product_name, tfidf_matrix)

print(f"\nProduits similaires à '{product_name}':")
print(similar_products)
```
